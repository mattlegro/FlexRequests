```python
def get_game_stats(appid):

    payload = {
    "appid" : appid,
    "key" : steam_key,
    "steamid": steam_id,
    'l' : 'en'    
    }

    web_api = 'ISteamUserStats'
    method = 'GetPlayerAchievements'
    version = 'v0001'

    url = f"http://api.steampowered.com/{web_api}/{method}/{version}/"
    ga_response = requests.get(url, params=payload)

    ga_json_obj = json.loads(ga_response.text)

    try:
        game_name = ga_json_obj['playerstats']['gameName']
        game_achievements = ga_json_obj['playerstats']['achievements']
        num_achievements = sum([achievement['achieved'] for achievement in game_achievements])
    except:
        game_name = None
        num_achievements = None

    return [ game_name, num_achievements ]
```

```python
get_game_stats(game_playtime_table['appid'][0])
```

```python
game_playtime_table[['game_name','game_achievements']] = game_playtime_table.apply(lambda x: get_game_stats(x['appid']), axis = 1, result_type='expand')
```
